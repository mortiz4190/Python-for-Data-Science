## To perform data analysis, we will slice out formatted dataframes form the original, we must: 

1). Remove irrelevant data: Filer out data or obs. that arent relevant to questions you want to answer.

2). Deduplicate data: Make sure there is no redundant information in dataframes.

3). Fix structural errors in data: Mispellings, incongruent naming conventions, incorrect word use, datetimes, etc. Make sure your columns are standardized for optimal queries.

4). Deal with missing data: Must deal with NANs in the three possible ways,
omission, fill in the blank data using approximations based on previous data point, average out datapoints above and below to give approximation.

5). Filter out data outliers (especially when it comes time to graph, will skew graphics significantly): Filter out data that may skew your findings, take an extra look at things that dont add up.

6). Validate your data: Authenticate your data, make sure its high quality, consistent, and properly formatted for downstream use.

        - Is it uniformly formatted in a design or language that your analysis tools can work with?
        - Does your clean data immediately prove or disprove your theory before analysis?

```python
# Fixes datetime to pandas format: year-month-date -> hour
df["datetime"] = df["datetime"].apply(lambda x: x.replace("24:00", "00:00")) #Changes from standard time to military time
df['datetime'] = pd.to_datetime(df['datetime']) #Changes format to year-month-date -> hour
```

```python
csc_duration = df.groupby(['datetime','city', 'state', 'country', 'shape'])['duration (seconds)'].sum().reset_index() #Now datetime is fixed to be in (year-month-date -> hour) format
csc_duration
```

#### Combine the City, State, Country text together in one strings to use it as the x label in the bar plot

```python
csc_duration['csc'] = csc_duration.city + ', ' + csc_duration.state + ', ' + csc_duration.country #This formats the location column to city, province or state, country
```

```python
csc_duration['hours'] = csc_duration['duration (seconds)'] / (60 * 60) #This formats the hours column to raw hours 
```

#### Sort values

```python
csc_duration = csc_duration.sort_values('duration (seconds)', ascending=False) #Sorts second duration from greatest to least
```

# UFO Sightings per Weekday

```python
csc_duration_weekday = csc_duration.groupby('weekday')[['city']].count() #Tells us which weekday had the most sightings
csc_duration_weekday = csc_duration_weekday.sort_values('city', ascending=False) #Sorting from greatest to least
csc_duration_weekday # Friday is the day with the most sightings which makes sense, it is when most are off from work, more likely to be outdoors.
```

```python
plt.title('UFO Sightings per Weekday')
plt.ylabel('Number of Sightings')
plt.xlabel('Day of the Week (0=Sunday, 6=Saturday)')
j = sns.scatterplot(x='weekday', y='city', data=csc_duration_weekday)
ax = sns.lineplot(x='weekday', y='city', data=csc_duration_weekday)
sns.set(rc={"figure.figsize":(6, 4)})
ax.set_xticks(range(0,7), labels=range(0, 7));
```

# UFO Sightings per Month

```python
csc_duration_month = csc_duration.groupby('month')[['city']].count() #Tells us which months had the most sightings
csc_duration_month = csc_duration_month.sort_values('city', ascending=False) #Sorting from greatest to least
csc_duration_month # July had the most sightings which makes sense, it is summer when more people are outdoors.
```

```python
#Plot is GOOD TO GO
plt.title('UFO Sightings per Month')
plt.ylabel('Number of Sightings')
plt.xlabel('Month (1 = January, 12 = December)')
j = sns.scatterplot(x='month', y='city', data=csc_duration_month)
ax = sns.lineplot(x='month', y='city', data=csc_duration_month)
ax.set_xticks(range(1,13), labels=range(1, 13));
```

# UFO Sightings by Shape

```python
csc_duration_shape = csc_duration.groupby('shape')[['city']].count()
csc_duration_shape = csc_duration_shape.sort_values('city', ascending=False)
csc_duration_shape.head(10)
```

```python
csc_duration_shape = csc_duration.groupby('shape', as_index=False)[['city']].count()
csc_duration_shape = csc_duration_shape.sort_values('city', ascending=False) #Sorting from greatest to least
s = csc_duration_shape[0:10] #Light is most common shape, followed by trianlge, then a circle
plt.title('Top 10 UFO Sightings by Shape')
sns.barplot(x='city',y='shape' ,data= s);
plt.ylabel('Different Shapes')
plt.xlabel('Number of Sightings')
```
